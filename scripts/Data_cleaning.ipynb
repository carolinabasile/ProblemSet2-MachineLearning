```python
# Descomprimo las bases
for file in os.listdir():
    if file.endswith('.rar'):
        !unrar x {file}


# Leo las bases
dataframes = []

for csv_file in os.listdir():
    if csv_file.endswith('.csv'):
        df = pd.read_csv(csv_file)
        dataframes.append(df)
        print(csv_file)

train_hogares = dataframes[0]
test_personas = dataframes[1]
test_hogares = dataframes[2]
train_personas = dataframes[3]
```

```python
# Uno las bases de hogares con las de individuos
train_df = pd.merge(train_hogares, train_personas, on='id', how='inner')

test_df = pd.merge(test_hogares, test_personas, on='id', how='inner')
```

```python
# Convierto las bases unidas a .csv

train_df.to_csv('/content/ProblemSet2-MachineLearning/stores/input/train_df.csv', index=False)
test_df.to_csv('/content/ProblemSet2-MachineLearning/stores/input/test_df.csv', index=False)
```

```python
#  Comprimo las bases unidas para poder exportarlas al repositorio

train_file = '/content/ProblemSet2-MachineLearning/stores/input/train_df.csv'
test_file = '/content/ProblemSet2-MachineLearning/stores/input/test_df.csv'
train_zip = '/content/ProblemSet2-MachineLearning/stores/train_df.zip'
test_zip = '/content/ProblemSet2-MachineLearning/stores/test_df.zip'

with ZipFile(train_zip, 'w') as zipf:
    zipf.write(train_file)
print(f"Created zip file: {train_zip}")

with ZipFile(test_zip, 'w') as zipf:
    zipf.write(test_file)
print(f"Created zip file: {test_zip}")
```
